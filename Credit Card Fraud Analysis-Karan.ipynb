Balancing the dataset

```python
X = newdata.drop('Class', axis=1)
Y = newdata['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score,precision_score
```

```python
smote = SMOTE(random_state=42) 
x_smote, y_smote = smote.fit_resample(X, Y)
```

```python
y_smote.value_counts()
```

```python
corr_balanced = x_smote.corr()
```

```python
sns.heatmap(corr_balanced)
```

```python
x_smote
```

```python
fraud_transaction = newdata[newdata['Class'] == 0]
real_transaction = newdata[newdata['Class'] == 1]
plt.figure(figsize=(20, 20))
plt.xlim(0, 6000)
plt.ylim(0, 5000)
plt.title("Fraud vs. Real Transactions")
plt.xlabel("Amount")
plt.ylabel("Frequency")
sns.histplot(fraud_transaction['Amount'],label="Fraud")
sns.histplot(real_transaction['Amount'],label="Real")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 20))
plt.title("Transaction Amount")
sns.boxplot(x="Class", y="Amount", data=newdata)
plt.show()
```

Logistic Regression Model

```python
X_train, X_test, Y_train, Y_test = train_test_split(x_smote, y_smote, test_size=0.20, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
y_pred_lr = lr.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred_lr)
confusion = confusion_matrix(Y_test, y_pred_lr)


print(f"Accuracy Score of Logistic Regression is : {accuracy}")
print(f"Confusion Matrix : \n{confusion}")
```

```python
X_train=pd.DataFrame(X_train)
```

```python
num_cols=X_train._get_numeric_data().columns
```

```python
X_train[num_cols].hist()
```

```python
confusion_matrix_logistic = confusion_matrix(Y_test, y_pred_lr)
classification_report_logistic = classification_report(Y_test, y_pred_lr, output_dict=True)

confusion_matrix_df = pd.DataFrame(confusion_matrix_logistic, 
                                   columns=["Predicted Negative (0)", "Predicted Positive (1)"],
                                   index=["Actual Negative (0)", "Actual Positive (1)"])

classification_report_df = pd.DataFrame(classification_report_logistic)

print("Logistic Regression Model - Confusion Matrix:")
print(confusion_matrix_df)

print("\nLogistic Regression Model - Classification Report:")
print(classification_report_df)
```

Hyperparameter Boosting Using XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(max_depth=13,n_estimators=300,learning_rate=0.5)

xgb.fit(X_train, Y_train)

y_pred_xgbc = xgb.predict(X_test)
```

```python
precision_score_xgb =  precision_score(Y_test, y_pred_xgbc)
accuracy_score_xgb = accuracy_score(Y_test, y_pred_xgbc)
print('The precision score  is = ',(precision_score_xgb * 100), '%')
print('The accuracy score is = ',(accuracy_score_xgb * 100), '%')
print (classification_report(Y_test,y_pred_xgbc))
```

Decision Tree - Model Validation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)

y_pred_dtc = dtc.predict(X_test)

acc_dtc = accuracy_score(Y_test, y_pred_dtc)
conf = confusion_matrix(Y_test, y_pred_dtc)
clf_report = classification_report(Y_test, y_pred_dtc)

print(f"Accuracy Score of Decision Tree is : {acc_dtc}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```
